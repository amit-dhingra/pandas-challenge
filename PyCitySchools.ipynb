```python
 # Dependencies and Setup
import pandas as pd

# Files to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
#school_data_complete
```

```python
number_of_schools= school_data_complete.school_name.nunique()
#print(number_of_schools)
```

```python
total_students = school_data_complete.student_name.count()
#print(total_students)
```

```python
total_budget = school_data['budget'].sum() 
#print(total_budget)
```

```python
sum_math_score = school_data_complete['math_score'].sum() 
average_math_score = sum_math_score/total_students
#print(average_math_score)
```

```python
sum_reading_score = school_data_complete['reading_score'].sum() 
average_reading_score = sum_reading_score/total_students
#print(average_reading_score)
```

```python
overall_pass_rate  = (average_math_score + average_reading_score)/2
#print(overall_pass_rate)
```

```python
sum_math_score_passing = school_data_complete.loc[school_data_complete['math_score']>=70,'math_score'].count() 
#print(sum_math_score_passing)
percent_math_passing = (sum_math_score_passing/total_students)*100
#print(percent_math_passing)
```

```python
sum_reading_score_passing = school_data_complete.loc[school_data_complete['reading_score']>=70,'reading_score'].count() 
#print(sum_math_score_passing)
percent_reading_passing = (sum_reading_score_passing/total_students)*100
#print(percent_reading_passing)
```

```python
School_Metrics = pd.DataFrame({'Total Schools': number_of_schools, 'Total Students': total_students,'Total Budget':total_budget,'Average Math Score':average_math_score,'Average Reading Score':average_reading_score,'% Passing Math':percent_math_passing,'% Passing Reading':percent_reading_passing,'% Overall Passing Rate':overall_pass_rate}, index=[0])
School_Metrics['Total Budget'] = School_Metrics['Total Budget'].map('${:,.2f}'.format)
School_Metrics
```

```python
per_student_budget = school_data['budget']/school_data['size']
#print(per_student_budget)
```

```python
school_key_metrics = school_data_complete.groupby('school_name').agg({"type":"first","size":"first","budget":"first","math_score":"mean","reading_score":"mean"})
school_key_metrics.rename(columns = {'type':'School Type', 'size':'Total Students', 'budget':'Total School Budget', 'math_score':'Average Math Score',
                              'reading_score':'Average Reading Score'}, inplace = True) 
school_key_metrics['Per Student Budget'] = school_key_metrics['Total School Budget']/school_key_metrics['Total Students']
school_key_metrics['passing_math'] = school_data_complete.groupby('school_name')['math_score'].apply(lambda x: x[x>=70].count())
school_key_metrics['% Passing Math'] =(school_key_metrics['passing_math']/school_key_metrics['Total Students'])*100
school_key_metrics['passing_reading'] = school_data_complete.groupby('school_name')['reading_score'].apply(lambda x: x[x>=70].count())
school_key_metrics['% Passing Reading'] =(school_key_metrics['passing_reading']/school_key_metrics['Total Students'])*100
school_key_metrics['% Overall Passing Rate'] =(school_key_metrics['% Passing Math']+school_key_metrics['% Passing Reading'])/2
school_key_metrics.drop(["passing_math", "passing_reading"], axis=1 ,inplace = True) 
school_key_metrics_copy = school_key_metrics[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate','Per Student Budget']]
school_key_metrics_copy_students = school_key_metrics[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate','Total Students']]
school_key_metrics_copy_school_type = school_key_metrics[['Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate','School Type']]
school_key_metrics['Total School Budget'] = school_key_metrics['Total School Budget'].map('${:,.2f}'.format)
school_key_metrics['Per Student Budget'] = school_key_metrics['Per Student Budget'].map('${:,.2f}'.format)
school_key_metrics = school_key_metrics[['School Type','Total Students','Total School Budget','Per Student Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing Rate']]
print("Top Performing Schools (By Passing Rate)")
school_key_metrics.sort_values(by='% Overall Passing Rate', ascending=False).head(5)

```

```python
ninth_grade_list = school_data_complete.loc[school_data_complete["grade"] == "9th"].groupby("school_name")
tenth_grade_list = school_data_complete.loc[school_data_complete["grade"] == "10th"].groupby("school_name")
eleventh_grade_list = school_data_complete.loc[school_data_complete["grade"] == "11th"].groupby("school_name")
twelfth_grade_list = school_data_complete.loc[school_data_complete["grade"] == "12th"].groupby("school_name")
ninth_math_avg_df = pd.DataFrame(ninth_grade_list["math_score"].mean())
tenth_math_avg_df = pd.DataFrame(tenth_grade_list["math_score"].mean())
eleventh_math_avg_df = pd.DataFrame(eleventh_grade_list["math_score"].mean())
twelfth_math_avg_df = pd.DataFrame(twelfth_grade_list["math_score"].mean())
avg_math_bygrade_df = pd.merge(ninth_math_avg_df, tenth_math_avg_df, on = "school_name", how = "inner")
avg_math_bygrade_df = pd.merge(avg_math_bygrade_df, eleventh_math_avg_df, on = "school_name", how = "inner")
avg_math_bygrade_df = pd.merge(avg_math_bygrade_df, twelfth_math_avg_df, on = "school_name", how = "inner")
avg_math_bygrade_df.columns = ["9th", "10th", "11th", "12th"]
avg_math_bygrade_df
```

```python
#reuse the lists created above
ninth_read_avg_df = pd.DataFrame(ninth_grade_list["reading_score"].mean())
tenth_read_avg_df = pd.DataFrame(tenth_grade_list["reading_score"].mean())
eleventh_read_avg_df = pd.DataFrame(eleventh_grade_list["reading_score"].mean())
twelfth_read_avg_df = pd.DataFrame(twelfth_grade_list["reading_score"].mean())
avg_read_bygrade_df = pd.merge(ninth_read_avg_df, tenth_read_avg_df, on = "school_name", how = "inner")
avg_read_bygrade_df = pd.merge(avg_read_bygrade_df, eleventh_read_avg_df, on = "school_name", how = "inner")
avg_read_bygrade_df = pd.merge(avg_read_bygrade_df, twelfth_read_avg_df, on = "school_name", how = "inner")
avg_read_bygrade_df.columns = ["9th", "10th", "11th", "12th"]
avg_read_bygrade_df
```
